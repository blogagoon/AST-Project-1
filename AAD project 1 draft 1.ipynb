```python
import pandas as pd

# I'll be using the astrophysical parameters section, specifically the first file. However, this file is massive.
# My goal will be formatting it so it's easier to work with, isolating the cepheid stars, and then trimming it down.

# There are 1,541 lines of pure commentary for the astrophysical parameters file, so I will skip those.
df2 = pd.read_csv('AP_proper.csv.gz', skiprows = 1541, compression='gzip')
df2.head()
```

```python
# This filter will get rid of any data that doesn't include luminosity.
nan_luminocity_filter = df2["lum_flame"].isna()
df2[~nan_luminocity_filter]
```

```python
# IDs of cepheids.
var_ceph = pd.read_csv('ceph_var.csv.gz', skiprows = 583, compression='gzip')
sourceid = var_ceph["source_id"]
sourceid
```

```python
# Trying to isolate the cepheids but is producing 0.
ceph_filter = df2["source_id"].isin(sourceid)
df2[ceph_filter]
```

```python
# A summary table for variable stars.
vari_sum = pd.read_csv('vari_sum.csv.gz', skiprows = 461, compression='gzip')
vari_sum.head()
```

```python
# Another attempt to isolate cepheid stats.
ceph_filter = vari_sum["in_vari_cepheid"]
ceph_source_ids = vari_sum[ceph_filter]["source_id"]
df2[df2["source_id"].isin(ceph_source_ids)]
```
