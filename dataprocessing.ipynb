```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pulsationdf = pd.read_csv('allcepheids.csv')
astrometrydf = pd.read_csv('cepheidastrometry.csv')
astrometrydf.rename(columns={'SOURCE_ID': 'source_id'}, inplace=True)
pulsationdf.head()
```

```python
combineddf = pd.merge(pulsationdf, astrometrydf, on='source_id')

def magconvert(appMag, parallax):
    absMag = appMag+5*np.log10(parallax)-10
    return absMag

combineddf['absMag'] = magconvert(combineddf.int_average_rp, combineddf.parallax)
print(combineddf.type_best_classification)
```
